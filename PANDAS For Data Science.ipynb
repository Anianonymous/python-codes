**Concat**

```python
# Read both files
df1 =pd.read_csv('df1.csv')
df2 =pd.read_csv('df2.csv')
```

```python
df3 = pd.concat([df1,df2],axis = 0)
df3.shape
```

```python
df4 = pd.concat([df1,df2],axis = 1)
df4.shape
```

**join / merge**

```python
# inner join 
df5 = pd.merge(df1, df2, on='RollNo', how='inner')
df5
```

```python
# Left outer
df6 = pd.merge(df1, df2, on='RollNo', how='left')
df6
```

```python
# Right Outer
df7 = pd.merge(df1, df2, on='RollNo', how='right')
df7
```

```python
# Full Outer
df8 = pd.merge(df1, df2, on='RollNo', how='outer')
df8
```

```python
#Replace null with one value
dsn = pd.read_csv('data_cleaning.csv')
dsn["Calories"].fillna(130, inplace = True)
pd.isnull(dsn).sum()
```

```python
#replace null with mean
dsn["Calories"].fillna(dsn["Calories"].min(), inplace = True)
pd.isnull(dsn).sum()
```

```python
#replace null with mean
dsn["MaxPulse"].fillna(dsn["MaxPulse"].mean(), inplace = True)
pd.isnull(dsn).sum()
```

```python
# replace with median
dsn["Pulse"].fillna(dsn["Pulse"].median(), inplace = True)
pd.isnull(dsn).sum()
```

```python
dsn.info()
```

```python
#converting to datetime format
dsn['Date'] = pd.to_datetime(dsn['Date'])
dsn.info()
```

```python
#Drop duplicates
print(dsn.duplicated())
```

```python
dsn.drop_duplicates(inplace = True)
print(dsn.duplicated())
```
